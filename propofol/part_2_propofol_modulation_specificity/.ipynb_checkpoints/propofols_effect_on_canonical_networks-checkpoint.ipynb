```python
import pandas as pd
import numpy as np

from propofol.utils.subjects import subjects
from propofol.utils.create_standard_network_heatmaps import create_standard_network_heatmaps, make_ramp

from fc_utils.get_edge_vals import get_edge_vals

from scipy import stats
import matplotlib.pyplot as plt

from fc_utils.load_fc import load_fc
from fc_utils.from_node_to_network import from_node_to_network, conv_to_network

import matplotlib
import seaborn as sns
import scipy.io as sio

import matlab.engine
import os

%load_ext autoreload
%autoreload 2
```

# load FC data, calculate network strength values for each canonical network 

```python
data_dir = 'censor_frames'

runs = ['01', '03']
tasks = ['rest', 'movie']
sessions = ['01']

data_dict = {
    'subject':[],
    'task':[],
    'sedation_level':[],
    'flat_mat_index':[],
    'sum_of_strengths':[],
}


nodes_to_drop = None
for sub in subjects:
    for session in sessions:
        for task in tasks:
            for run in runs:
                path = f'../data/{data_dir}/{sub}_{session}_{task}_{run}_LPI_000.netcc'
                fc = None
                if os.path.exists(path):
                    fc = load_fc(path,
                                 nodes_to_drop=nodes_to_drop)
                if fc is not None:
                    net_mat, labels = conv_to_network(fc, func='avg')

                    net_mat = get_edge_vals(net_mat, exclude_diag=False)
                    for flat_mat_index in range(36):
                        data_dict['flat_mat_index'].append(flat_mat_index)
                        data_dict['subject'].append(sub)
                        data_dict['sedation_level'].append(run)
                        data_dict['task'].append(task)
                        data_dict['sum_of_strengths'].append(net_mat[flat_mat_index])
                else:
                    print('didnt load" ', sub, task, run)
```

# turn all fo this data into df, and drop any scans with fraction censored over .5

```python
df = pd.DataFrame(data_dict)
motion_df = pd.read_csv('../data/ss_out_review_compiled/ss_out_review_compiled_censor_frames.csv')
motion_df['subject'] = motion_df['sub']
motion_df['sedation_level'] = motion_df['run'].apply(lambda x: "0" + str(x))
df = pd.merge(df, motion_df, on=['subject', 'sedation_level', 'task'])
df = df[df['fraction censored per run']<.5]
```

```python
def make_small_heatmap(heatmap, 
                       p_val_labels = [['*', '*'],['*', '*'],],
                       yticklabels=['high-attention network','low-attention network'], 
                       color_range=(-4.24, 2.90)):
    sns.set_style("white")
    sns.set_context("notebook", font_scale=2.6, rc={"lines.linewidth": 2.5, 'font.family':    'serif',
                                                  'font.serif':'Helvetica'})
    fig, ax = plt.subplots(1,  figsize=(2,2), dpi=150)

    cmap = make_ramp( ['#54BABA','#FFFFFF',  '#E59E37'] )
    sns.heatmap(heatmap, 
                    ax = ax,
                    yticklabels=yticklabels,
                    cmap = sns.diverging_palette(189, 37, s=100, l=66, as_cmap=True),
                    vmin=color_range[0], vmax=color_range[1],
                    annot=p_val_labels,
                    fmt="s",
                    cbar=False,
                    cbar_ax = False,
                   linewidths=4)
    ax.tick_params(
    axis='x',         
    which='both',
        labelbottom=False,
    left=False,)   


    for _, spine in ax.spines.items():
        spine.set_visible(True)  
    return fig
```
